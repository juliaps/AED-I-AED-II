```python
import random 
import math

def dinamica_molecular(m, tam, N ,u0, r0):
    
    sx = []
    sy = []
    sz = []
    vx = []
    vy = []
    vz = []
    ax = []
    ay = []
    az = []
    
    Fx = 0.0
    Fy = 0.0
    Fz = 0.0
    Fr = 0.0
    vux = 0.0
    vuy = 0.0
    vuz = 0.0
    dij = 0.0
    dx = 0.0
    dy = 0.0
    dz = 0.0
    dt = 0.01
    
    for i in range (0, N):
        
        sx.append(random.uniform(0,tam))
        sy.append(random.uniform(0,tam))
        sz.append(random.uniform(0,tam))
        vx.append(random.uniform(0,tam))
        vy.append(random.uniform(0,tam))
        vz.append(random.uniform(0,tam))
        ax.append(random.uniform(0,tam))
        ay.append(random.uniform(0,tam))
        az.append(random.uniform(0,tam))
    
    for i in range (0,N):
        
        Ftx = 0.0
        Fty = 0.0
        Ftz = 0.0
        
        for j in range (0, N):
            
            if i!=j:
                dij = math.sqrt((sx[j]-sx[i])**2 + (sy[j] - sy[j])**2 + (sz[j]-sz[i])**2)
                vux = (sx[j] - sx[i])/dij
                vuy = (sy[j] - sy[i])/dij
                vuz = (sz[j] - sz[i])/dij
                
                Fr = 12 * u0/r0 * (r0/dij)**13 - (r0/dij)**7
                
                Fx = Fr*vux
                Fy = Fr*vuy
                Fz = Fr*vuz
                
                Ftx = Ftx + Fx
                Fty = Fty+ Fy
                Ftz = Ftz + Fz
                
        ax[i] = Ftx/m
        ay[i] = Fty/m
        az[i] = Ftz/m
        
        vx[i] = vx[i] + ax[i]*dt
        vy[i] = vy[i] + ay[i]*dt
        vz[i] = vz[i] + az[i]*dt
        
        dx = vx[i]*dt
        dy = vy[i]*dt
        dz = vz[i]*dt
        
        if (sx[i]+dx)< 0 or (sx[i]+dx)>tam:
            dx = -dx
        
        if (sy[i]+dy)< 0 or (sy[i]+dy)>tam:
            dy = -dy
        
        if (sz[i]+dz)< 0 or (sz[i]+dz)>tam:
            dz = -dz
            
        sx[i] = sx[i] + dx
        sy[i] = sy[i] + dy
        sz[i] = sz[i] + dz
        
    return sx,sy,sz
        
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
T = 100
tempo = []
mx = []
my = []
mz = []

for i in range (0,T):
    sx, sy, sz = dinamica_molecular(1,10,100,1,1)
    mx.append(sx)
    my.append(sy)
    tempo.append(i)
    
plt.figure(figsize=(20,10))
plt.plot(tempo, mx, label = "Posição no eixo X")
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
```

```python
plt.figure(figsize=(20,10))
plt.plot(tempo, my, label ="Posição no eixo Y")
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
```
