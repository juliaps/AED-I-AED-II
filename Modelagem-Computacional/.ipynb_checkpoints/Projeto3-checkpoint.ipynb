## Projeto 3 : Sincronização de neuronios



Inicialmente foi importado as bibliotecas para a implementação dos relaxamentos e geração dos grafos, e definida as constantes.                                
                                      

                                      
                                      
                                      

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import random
n = 500
m = 5
```

Posteriormente foi implementada a função de relaxamento com acoplamento, tendo como entrada o grafo,o numero de neuronios, o tempo maximo e p(ro) embora p nesses casos sejam iguais a zero.

```python
def relaxamento_acoplamento(grafo, N, TMAX, p):
    
    X = []
    T = []
    dt = 0.01
    epsilon = 0.02 
    alfa = 6.0
    beta = 0.1
    teta = 0.5
    w = 0.1
    I = 0.2
    H = 0
    t = 0
    count = 0
  
    x = [((float(random.randint(0, 1000)/1000.0)*4.0) - 2.0) for i in range(0, N)]
    y = [( float(random.randint(0, 1000)/1000.0)*4.0 ) for i in range(0, N)]
    S = [0 for i in range(0, N)]
    
    while (t < TMAX):
    
    
        for i in range(0, N):
            S[i] = 0
            for j in range(0, N):
                if grafo.has_edge(j,i) == True:
                    if x[j] - teta >= 0:
                        H = 1 
                    else:
                        H = 0   
                    S[i] = S[i] + w * H
                    
        for i in range(0, N):
            dx = (3*x[i] - x[i]**3 + 2 - y[i] + I + p + S[i])*dt
            dy = (alfa*(1 + np.tanh(x[i]/beta)) - y[i])*epsilon*dt

            x[i] = x[i] + dx
            y[i] = y[i] + dy

        t = t + dt
        
        if count % 100 == 0:
            T.append(t)
            X.append(x[:])
        
        count = count + 1
    return T, X
```

O primeiro teste a ser realizado é com o grafo regular, pelo gráfico podemos visualizar que em T = 1750 os neuronios foram sincronizados.

```python
g1 = nx.star_graph(n)
T, X = relaxamento_acoplamento(g1, 500, 2000, 0)
plt.figure(figsize=(20,10))
plt.plot(T, X)
```

No segundo teste com o grafo aleatório Erdos Renyi, os neuronios sincronizaram no primeiro instante.

```python
g2 = nx.erdos_renyi_graph(n,100)
T, X = relaxamento_acoplamento(g2, 500, 3000, 0)
plt.figure(figsize=(20,10))
plt.plot(T, X)
```

No terceiro teste , com o grafo de Barabasi Albert podemos ver no gráfico que os neuronios sicronizaram com T = 1250.

```python
g3 = nx.barabasi_albert_graph(n,m)
T, X = relaxamento_acoplamento(g3, 500, 2000, 0)
plt.figure(figsize=(20,10))
plt.plot(T, X)
```

Posteriormente foi feita algumas alterações no código para gerar uma seleção aleatória de neuronios inativos, mas tendo como entrada o percentual de quantos estarão inativos.

```python
def relaxamento_acoplamento_inativos(grafo, N, TMAX, p, percentual):
    
    X = []
    T = []
    dt = 0.01
    epsilon = 0.02 
    alfa = 6.0
    beta = 0.1
    teta = 0.5
    w = 0.1
    I = 0.2
    H = 0
    t = 0
    count = 0
    count_p = 0
    
    x = [((float(random.randint(0, 1000)/1000.0)*4.0) - 2.0) for i in range(0, N)]
    y = [( float(random.randint(0, 1000)/1000.0)*4.0 ) for i in range(0, N)]
    S = [0 for i in range(0, N)]
    Nr = [1 for i in range(0, N)]
    
    while (t < TMAX):
    
    
        for i in range(0, N):
            S[i] = 0
            if count_p < ((percentual*500)/100):
                a = random.randint(0,1)
                if a == 0:
                    Nr[i] = -0.1
                    count_p = count_p + 1
            
            for j in range(0, N):
                if grafo.has_edge(j,i) == True:
                    if x[j] - teta >= 0:
                        H = 1 
                    else:
                        H = 0   
                    S[i] = S[i] + w * H
                    
        for i in range(0, N):
            I = 0.2 * Nr[i]
            dx = (3*x[i] - x[i]**3 + 2 - y[i] + I + p + S[i])*dt
            dy = (alfa*(1 + np.tanh(x[i]/beta)) - y[i])*epsilon*dt

            x[i] = x[i] + dx
            y[i] = y[i] + dy

        t = t + dt
        
        if count % 100 == 0:
            T.append(t)
            X.append(x[:])
        
        count = count + 1
    return T, X
```

No primeiro teste, com grafo regular, com 85 % dos neuronios inativos no tempo igual a 1625, a maioria dos neuronios foram sincronizados.

```python
g1 = nx.star_graph(n)
T, X = relaxamento_acoplamento_inativos(g1, 500, 2000, 0, 85)
plt.figure(figsize=(20,10))
plt.plot(T, X)
```

Novamente no segundo teste, com grafo aleatório, os neuronios sincronizaram no primeiro instante e o gráfico teve esse comportamento inesperado, mesmo que a quantidade do grau esperado tenha sido alterado de 100 para 5.

```python
g2 = nx.erdos_renyi_graph(n,m)
T, X = relaxamento_acoplamento_inativos(g2, 500, 3000, 0, 25)
plt.figure(figsize=(20,10))
plt.plot(T, X)
```

Agora testando com o grafo barabasi albert, também houve alguns neuronios que não sincronizaram, mas a maioria sincronizou no tempo por volta de 2000, e com 75% de neuronios inativos.

```python
g3 = nx.barabasi_albert_graph(n,m)
T, X = relaxamento_acoplamento_inativos(g3, 500, 2500, 0, 75)
plt.figure(figsize=(20,10))
plt.plot(T, X)
```

Para realizar a selção guiada de neuronios inativos, foi selecionado um numero igual a uma porcentagem definida como parametro, onde os que compõem esse numero de neuronios inativos, são os neuronios com maior grau.

```python
def relaxamento_acoplamento_inativos_maior_grau(grafo, N, TMAX, p, percentual):
    
    X = []
    T = []
    dt = 0.01
    epsilon = 0.02 
    alfa = 6.0
    beta = 0.1
    teta = 0.5
    w = 0.1
    I = 0.2
    H = 0
    t = 0
    flag = 0
    count = 0
    count_p = 0
    l = 0 
    n_inativos = int((percentual * N)/ 100)
    x = [((float(random.randint(0, 1000)/1000.0)*4.0) - 2.0) for i in range(0, N)]
    y = [( float(random.randint(0, 1000)/1000.0)*4.0 ) for i in range(0, N)]
    S = [0 for i in range(0, N)]
    Nr = [1 for i in range(0, N)]
    m_grau = [-1 for i in range (0,n_inativos)]
    grau = [0 for i in range (0,N)]
    
    while (t < TMAX):
    
    
        for i in range(0, N):
            S[i] = 0
            
            grau[i] = grafo.degree(i)
            while flag and l < n_inativos == 0:
                if m_grau[l] == -1:
                    m_grau[l] = i
                    flag = 1
                elif grau[i] > grau[m_grau[l]]:
                    aux = m_grau[l]
                    for k in range(l+1,n_inativos):
                        aux2 = m_grau[k]
                        m_grau[k] = aux
                        aux = aux2
                    m_grau[l] = grau[i]
                    flag = 1
                l = l + 1
            flag = 0

            for j in range(0, N):
                if grafo.has_edge(j,i) == True:
                    if x[j] - teta >= 0:
                        H = 1 
                    else:
                        H = 0   
                    S[i] = S[i] + w * H
        for i in range(0,n_inativos):
            Nr[m_grau[i]] = -0.1
        for i in range(0, N):
            I = 0.2 * Nr[i]
            dx = (3*x[i] - x[i]**3 + 2 - y[i] + I + p + S[i])*dt
            dy = (alfa*(1 + np.tanh(x[i]/beta)) - y[i])*epsilon*dt

            x[i] = x[i] + dx
            y[i] = y[i] + dy

        t = t + dt
        
        if count % 100 == 0:
            T.append(t)
            X.append(x[:])
        
        count = count + 1
    return T, X
```

Foi feito apenas o teste com o grafo barabasi albert, e no tempo próximo de 1500 e com 5% de neuronios inativos, os neuronios sincronizaram.

```python
g3 = nx.barabasi_albert_graph(n,m)
T, X = relaxamento_acoplamento_inativos_maior_grau(g3, 500, 2500, 0, 5)
plt.figure(figsize=(20,10))
plt.plot(T, X)
```

Conclusão

Os tipos de gerações de rede não teve uma diferença muito discrepante exceto pelo grafo aleatório Erdos-Renyi, que apresentou um comportamento inesperado. Mas no primeiro caso, onde todos os neuronios estavam ativos, o grafo de Barabasi Albert teve melhor comportamento, em relação ao grafo regular sincronizando em menos tempo. Já no segundo caso, dos neuronios inativos selecionados de forma aleatória, o grafo regular teve melhor desemepenho, tanto pelo menor número de neuronios ativos sendo um total de 15% quanto pelo tempo para sincronizá-los. O último teste foi feito apenas com o grafo Barabasi Albert, que acabou demorando menos tempo que o de inativos aleátorios, onde se tinha apenas 25% ativos e no de seleção por maior grau, 95% ativos, e demorou mais tempo, 1500, em relação ao relaxamento com acoplamento com todod os neuronios ativos, 1250. Com isso podemos concluir que quanto menos neuronios ativos leva mais tempo para sincronizá-los.
