Neste projeto foi escolhido o tema de propagação de rumores, onde os estados para as pessoas definidos foram: crente, descrente e desinformada. Onde como entrada da função é determinado o tempo de execução, a quantidade de nós do grafo, e quantidade inicial de pessoas informadas. 

Quando mais 50% dos vizinhos que já são iformados de uma pessoa serem crentes, a pessoa passa a acreditar, caso contrário ela torna-se descrente. Já para a pessoa que está descrente se tornar crente é necessário que ao todo 70% ou mais de seus vizinhos sejam crentes. Já para deixar de acreditar só se 75% de seus vizinhos sejam descrentes.

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import random
n = 500
m = 2
```

```python
def propagacao_rumor(grafo, N, Ni, TMax):
    
    
    qtde_C = 0
    qtde_D = 0
    qtde_DI = 0
    p_C = 0
    p_D = 0
    p_DI = 0
    t_C = []
    t_D = []
    t_DI = []
    c_C = 0
    c_D = 0
    c_DI = N - Ni
    t = 0
    Tempo = []
    nx.set_node_attributes(grafo, "desinformado", "estado")
    
    for i in range (0,N):
        if t < Ni:
            if random.randint(0,1) == 1:
                grafo.nodes[i]["estado"] = "crente"
                c_C = c_C + 1
            else:
                grafo.nodes[i]["estado"] = "descrente"
                c_D = c_D + 1
            t = t + 1
        else:
            break
    
    t_C.append(c_C)
    t_D.append(c_D)
    t_DI.append(c_DI)
    Tempo.append(t)
            
    while(t< TMax):
        
        for i in range (0,N):
            lista  = grafo.neighbors(i)
            for j in lista:
                if grafo.nodes[j]["estado"] == "crente":
                    qtde_C = qtde_C + 1
                elif grafo.nodes[j]["estado"] == "descrente" :
                    qtde_D = qtde_D + 1
                else:
                    qtde_DI = qtde_DI + 1
                    
            p_C = (100*qtde_C)/(qtde_C + qtde_D)
            p_C_T = (100*qtde_C)/(qtde_C + qtde_D + qtde_DI) 
            p_D = (100*qtde_D)/(qtde_C + qtde_D + qtde_DI)
            p_DI = (100*qtde_DI)/(qtde_C + qtde_D + qtde_DI)
            
            qtde_C = 0
            qtde_D = 0
            qtde_DI = 0
            
            if grafo.nodes[i]["estado"] == "desinformado":
                if p_C >= 50:
                    grafo.nodes[i]["estado"] = "crente"
                    c_C = c_C + 1
                    c_DI = c_DI - 1
                else:
                    grafo.nodes[i]["estado"] = "descrente"
                    c_D = c_D + 1
                    c_DI = c_DI - 1
            elif grafo.nodes[i]["estado"] == "descrente":
                if p_C_T > 70:
                    grafo.nodes[i]["estado"] = "crente"
                    c_C = c_C + 1
                    c_D = c_D - 1
                else:
                    c_D = c_D + 1
            else:
                if p_D > 75:
                    grafo.nodes[i]["estado"] = "descrente"
                    c_D = c_D + 1
                    c_C = c_C - 1
                else:
                    c_C = c_C + 1
        
        t = t + 1
        t_C.append(c_C)
        t_D.append(c_D)
        t_DI.append(c_DI)
        Tempo.append(t)
        c_C = 0
        c_D = 0
        
    return Tempo, t_C, t_D, t_DI
```

```python
g1 = nx.barabasi_albert_graph(n,m)
Tempo, t_C, t_D, t_DI = propagacao_rumor(g1,n,250, 800)
plt.figure(figsize=(20,10))
plt.plot(Tempo, t_C, label = "crente")
plt.plot(Tempo, t_D, label = "descrente")
plt.plot(Tempo, t_DI, label = "desinformado")
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
```

Podemos visualizar que utilizando o grafo Barabasi Elbert, a maioria, cerca de 380 das pessoas se tornaram crentes, tendo uma pequena variação no instante aproximandamente de 250, onde todas as pessoas se tornam informadas.

```python
g2 = nx.erdos_renyi_graph(n,m)
Tempo, t_C, t_D, t_DI = propagacao_rumor(g2,n,250, 800)
plt.figure(figsize=(20,10))
plt.plot(Tempo, t_C, label = "crente")
plt.plot(Tempo, t_D, label = "descrente")
plt.plot(Tempo, t_DI, label = "desinformado")
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
```

Já no caso do grafo Erdos Renyi todas as pessoas se tornarama informadas ecrentes aproximadamente ao t = 250.

```python
g3 = nx.barbell_graph(m,n)
Tempo, t_C, t_D, t_DI = propagacao_rumor(g3,n,250, 800)
plt.figure(figsize=(20,10))
plt.plot(Tempo, t_C, label = "crente")
plt.plot(Tempo, t_D, label = "descrente")
plt.plot(Tempo, t_DI, label = "desinformado")
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
```

E por fim no grafo regular, a maioria,cerca de 400, das pessoas se tornaram descrentes também aproximandamente ao t = 250, onde também todas pessoas ficaram informadas, ou seja, cerca de 100 pessoas descrente.

### Conclusão

Através dos gráficos e analise dos resultados, notamos que mesmo usando os mesmos paramêtros apenas alterando as tipologias de rede, a estabilidade foi atingida no mesmo período. Entretanto podemos observar que a topologia teve grande influência nos resultados, já que em cada um deles foi obtido um resultado difernte. No caso Barabasi Albert, foi obtida a maioria crente. Já no Erdos Renyi foi obtido o consenso, ou seja, todas as pessoas acreditavam no rumor. E por fim na topologia regular 80% das pessoas se tornaram descrentes.
