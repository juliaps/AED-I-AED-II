Primeiramente foi feita a dedução das fórmulas (Vx, Vy) . Para definir a velocidade no eixo x, foi aplicada a 2 ª Lei de Newton, onde a força resultante é igual a força de atrito, então : 
$$F = m . a$$
$$ -k . vx = m . a$$
$$\frac{-k.Vx}{m}  = a$$
$$\frac{-k.Vx}{m} = \frac{dVx}{dt}$$
$$ (\frac{-k.Vx}{m}) dt = dVx$$
                                                         
Para determinar a velocidade no eixo y semelhante foi aplicada a 2ª Lei de Newton, onde a força resultante é igual a soma das forças peso e a de atrito, logo : 

$$ F = Fa + P$$ $$m .a = -k.Vy -m.g$$ $$a = \frac{-k.Vy}{m}  -g$$
$$\frac{dVy}{dt} = \frac{(-k.Vy)}{m} -g$$
$$dVy = (\frac{k.Vy}{m} + g) (-dt)$$

Onde a velocidade (V), e as posições referente aos eixos x e y, e as energias cinética, pontencial e total é dada pelas seguintes fórmulas : 

$$V^2 =Vx^2 + Vy^2$$

$$ dSx =  dSx  + Vx(t) dt + \frac{dVx dt}{2}$$

$$ dSy = dSy + Vy(t) dt + \frac{dVy dt}{2}$$


$$Ec = \frac{mV^2}{2}$$ 
$$Ep = m g h$$
$$Et = Ec + Ep$$

E para calcular as velocidades no eixos x e y, e por consequencia os demais elementos foi impletado o método de Euler na função lancamentoObliquo no código abaixo, onde o parametro é a constante de atrito.

```
# Júlia dos Passos Saraiva   RA : 103228
# Modelagem Computacional

import numpy as np
import matplotlib.pyplot as plt

def lancamentoObliquo(k):
  #Declaração dos parametros
  V  = 100
  teta = (60/180)*np.pi
  g = 10
  Vx = V * np.cos(teta)
  Vy = V * np.sin(teta)
  m = 3
  dt = 0
  vVx = []
  vVy = []
  deltaT = []
  deltaSx = []
  deltaSy = []
  Ec = []
  Ep = []
  Et = []
  dSy = 0
  dSx = 0
  flag = 1
  
  #método de Euler
  while(flag):
      deltaT.append(dt)

      vVx.append(Vx)
      deltaSx.append(dSx)

      vVy.append(Vy)
      deltaSy.append(dSy)

      dVx = (-1)*(k*Vx/m) * dt
      dVy = ((-1) * (g +(k*Vy/m))) *dt

      dSx = dSx + Vx * dt + dVx*dt/2
      dSy = dSy + Vy * dt + dVy*dt/2

      Vx = Vx + dVx
      Vy = Vy + dVy

      Ec.append((m/2) *((Vx * Vx) + (Vy * Vy)))
      Ep.append((m*g*dSy))
      Et.append((Ec[-1] + Ep[-1]))

      if dSy < 0:
          flag = 0

      dt = dt + 0.01
  
  return (vVx, vVy, deltaT, deltaSx, deltaSy, Ec, Ep, Et)
```

#Simulação com k = 1

###Gráfico da Velocidade  em função do tempo no eixo x e y

```
vVx, vVy , deltaT, deltaSx, deltaSy, Ec, Ep, Et = lancamentoObliquo(1)
plt.plot(deltaT, vVy, label="Vy")
plt.plot(deltaT, vVx, label="Vx")
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
```

###Energias em relação ao tempo

```
plt.plot(deltaT, Ec, label="Energia Cinética")
plt.plot(deltaT, Ep, label="Energia Potencial")
plt.plot(deltaT, Et, label="Energia Total")
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
```

###Altura e Posição em relação ao tempo

```
plt.plot(deltaT, deltaSy, label="Altura")
plt.plot(deltaT, deltaSx, label="Posição x")
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
```

#Conclusão 

Através dos gráficos podemos perceber que com diferentes constantes de atrito o comportamento continua o mesmo. Entretanto o alcance tanto da altura e da posição x são inversamente proporcionais a constante de atrito. E o momento em que altura máxima é atingido varia . A energia potencial depende da altura, então da mesma forma quando a constante de atrito aumenta menor a energia pontencial. Já a energia cinética depende da velocidade e como podemos observar nos gráficos com o aumento da constante de atrito a velocidade diminui mais rapidamente, principalmente no eixo y, onde a velocidade se torna negativa após o objeto atingir a sua altura, ou seja, quando ele começa cair. Logo a energia cinética também é inversamente proporcional a constante de atrito.
