Os algoritmos de Substituição de página são usados  no gerenciamento de memória para fazer o controle de quais elementos presentes na memória virtual devem ser inseridos na memória física.

Nesse Prática Foi desenvolvido dois desses algoritmos o FIFO e o LRU.
O Algoritmo Fifo, mantém uma fila das páginas correntes na memória.
Ele considera que a página no início da fila é a mais antiga e a página no final é a mais nova.
Quando ocorre um page fault:

 -A página do início é removida
 
 -A nova é inserida ao final da fila
 
 
O algortimo imprime a quantidade de page fault.

```python
import matplotlib.pyplot as plt

def fifo(a,n,m):
    f = -1
    page_faults = 0
    page = []
    for i in range(m):
        page.append(-1) ## preenchendo tudo de zero
    
    for i in range(n):
        flag = 0
        for j in range(m):
            if(page[j] == a[i]):
                flag = 1
                break

        if flag == 0:
            f=(f+1)%m
            page[f] = a[i]
            page_faults+=1         
    return page_faults
```

O outro algoritmo implementado foi o LRU, que considera que as páginas usadas com mais frequência nas ultimas
execuções provavelmente serão muito utilizadas novamente.
Assim se existir page fault o algoritmo elimina a página não utilizada pelo período de tempo mais longo.

```python
def lru(a,n,m):
    x = 0
    page_faults = 0
    page = []
    for i in range(m):
        page.append(-1)
    for i in range(n):
        flag = 0
        for j in range(m):
            if(page[j] == a[i]):
                flag = 1 ## sinaliza 1
                break
            
        if flag == 0:
            if page[x] != -1:
                min = 999
                for k in range(m):
                    flag = 0
                    j =  i
                    while j>=0:
                        j-=1
                        if(page[k] == a[j]):
                            flag = 1
                            break
                    if (flag == 1 and min > j):
                        min = j
                        x = k

            page[x] = a[i]
            x=(x+1)%m
            page_faults+=1
    return page_faults
```

Foi realizadas simulações de páginas para serem inseridas gerenciadas de acordo com cada algoritmo, e foi feita a análise de acordo com a quantidade de memória virtual.

```python
a= [7, 4, 2, 0, 1, 3 ,5, 4, 1, 0, 3, 1,5]
n=13

j=[]
k=[]
l= [] 
for x in range(1,9):
    j.append(fifo(a,n,x))
    k.append(lru(a,n,x))
    l.append(x)
    

plt.figure(figsize=(20,10))
plt.plot(l,k,label="Fifo")
plt.plot(l,j,label="LRU")
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)


```

Pelo gráfico é possível analisar  que o Algoritmo FIFO possui mais Page Fauls que o algoritmo LRU, no intervalo em que a memória virtual pode armazenar de 3 a 5 páginas.
Por esses testes pode-se concluir que o LRU é melhor que o FIFO nesses casos, mas nos demais eles tem o mesmo desempenho.
